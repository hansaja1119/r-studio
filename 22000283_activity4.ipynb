```r
library(ggplot2)
library(moments)
```

```r
heart_data <- read.csv("heart.csv") 
head(heart_data)
```

```r
ggplot(heart_data, aes(x = as.factor(Sex), y = MaxHR, fill = as.factor(Sex))) +
  geom_boxplot() +
  scale_fill_manual(values = c("purple", "yellow")) + 
  labs(x = "Sex", y = "Maximum Heart Rate", title = "Heart Rates") +
  theme_minimal()
```

```r
boxplot_stats <- boxplot(heart_data$MaxHR ~ heart_data$Sex, plot = FALSE)
boxplot_stats$out  
```

```r
male_skewness <- skewness(heart_data$MaxHR[heart_data$Sex == 'M'])
female_skewness <- skewness(heart_data$MaxHR[heart_data$Sex == 'F'])

print(paste("Skewness for Male:", male_skewness))
print(paste("Skewness for Female:", female_skewness))
```

```r
chess_data <- read.csv("games.csv")  
```

```r
ggplot(chess_data, aes(x = winner, y = turns, fill = winner)) +
  geom_boxplot() +
  scale_fill_manual(values = c("green", "blue", "red")) + 
  labs(x = "Winner of the game", y = "No of turns", title = "Chess game summary") +
  theme_minimal()
```

```r
boxplot_stats <- boxplot(chess_data$turns ~ chess_data$winner, plot = FALSE)
boxplot_stats$out 
```

```r
skew_winner <- tapply(chess_data$turns, chess_data$winner, skewness)
print(skew_winner)
```

```r
set.seed(123)
mydata1 <- rnorm(100, mean = 50, sd = 10) 

boxplot(mydata1, col = "lightblue", main = "Random Data Boxplot", ylab = "Values")
```

```r
summary(mydata1)
```

```r
range_value <- max(mydata1) - min(mydata1)
print(paste("Range:", range_value))
```

```r
skew_value <- skewness(mydata1)
print(paste("Skewness:", skew_value))
```

```r
boxplot_stats <- boxplot.stats(mydata1)
boxplot_stats$out  
```
