```r
movie_types <- c("Comedy", "Action", "Romance", "Drama", "SiFi")
```

```r
no_of_likes <- c(40, 50, 60, 10, 30)
```

```r
pie(no_of_likes, labels = movie_types, col = rainbow(length(movie_types)), main = "Favorite type of Movie")
```

```r
percent_labels  <- round(no_of_likes / sum(no_of_likes) * 100, 1)
percent_labels <- paste(percent_labels, "%", sep="") 
```

```r
pie(no_of_likes, labels = percent_labels, col = rainbow(length(movie_types)), main = "Favorite type of Movie")
```

```r
barplot(no_of_likes, 
        names.arg = movie_types, 
        col = "red", 
        border = "yellow", 
        main = "Favourite type of Movie", 
        xlab = "Movie Type", 
        ylab = "No of Likes")
```
